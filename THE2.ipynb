```python
from tkinter import *
import tkinter

window=Tk()
window.resizable(0, 0)
window.configure(background="light blue")
window.geometry("300x300")
window.title("Pandemic Analytics Engine")

def cal_func():

    total_population = entry2.get()
    int_total_population = int(total_population)

    infected = entry3.get()
    int_infected = int(infected)

    death = entry4.get()
    int_death = int(death)

    ifr = (int_death / int_infected) * 100
    cmr = (int_death / int_total_population)*10000
    
    window = Tk()
    window.title("Pandemic Analyzer")
    window.configure(background="green")
    window.geometry("300x300")
    
    
    
    label1 = Label(window, font=("Arial", 22, "bold"), text="COVID19 disease" , bg = "green" , fg = "blue")
    label1.grid(row=0, column=0)

    label2 = Label(window, font=("Arial", 18, "bold"), text="CMR : ", bg = "green" , fg = "blue")
    label2.grid(row=2, column=0)
   
    label3 = Label(window, text=cmr)
    label3.grid(row=2, column=1)

    label4 = Label(window, font=("Arial", 18, "bold"), text="IFR : ", bg = "green" , fg = "blue")
    label4.grid(row=3, column=0)
    
    label5 = Label(window, text=ifr)
    label5.grid(row=3, column=1)
    label6= Label(window, font=("Arial", 8),text="(deaths per 10000 population in state)" ,bg="green", fg = "blue")
    label6.grid(row=5,column=0)
       
options = ['Andhra Pradesh', 'Arunachal Pradesh', 'Assam', 'Bihar', 'Chandigarh', 'Chhattisgarh', 'Dadra and Nagar Haveli', 'Delhi', 'Goa', 'Gujarat', 'Haryana', 'Himachal Pradesh', 'Jammu and Kashmir', 'Jharkhand', 'Karnataka', 'Kerala', 'Ladakh', 'Lakshadweep', 'Madhya Pradesh', 'Maharashtra', 'Manipur', 'Meghalaya', 'Mizoram', 'Nagaland', 'Odisha', 'Puducherry', 'Punjab', 'Rajasthan', 'Sikkim', 'Tamil Nadu', 'Telengana', 'Tripura', 'Uttar Pradesh', 'Uttarakhand', 'West Bengal']
select = StringVar() 
select.set(options[0])

label1 = Label(window,font=("Arial", 22, "bold"), text="Pandemic Analyzer Calculator",bg = "red" , fg = "blue")
label1.grid(row=0,column= 1)

label2 = Label(window, text="Select State:",bg = "pink" , fg = "blue")
label2.grid(row=2,column=0)
drop = OptionMenu(window,select,*options)
drop.grid(row=2,column=1)

label3 = Label(window, text="Infected:" ,bg="red" , fg = "blue")
label3.grid(row=3,column=0)
entry2 = Entry(window)
entry2.grid(row=3, column=1)

label4 = Label(window, text="Recoverd:" ,bg="brown", fg = "blue")
label4.grid(row=4,column=0)
entry3 = Entry(window)
entry3.grid(row=4, column=1)

label5 = Label(window, text="Deceased:" ,bg="grey", fg = "blue")
label5.grid(row=5,column=0)
entry4 = Entry(window)
entry4.grid(row=5, column=1)

label6 = Label(window, font=("Arial", 12, "italic"),text="CMR" ,bg="black", fg = "blue")
label6.grid(row=10,column=0)
label7= Label(window, font=("Arial", 12, "italic"),text="IFR" ,bg="black", fg = "dark blue")
label7.grid(row=15,column=0)

button1 = Button(window, text="   CMR    ",command=cal_func , bg = "grey" , fg = "blue")
button1.grid(row=7,column=1)
button1 = Button(window, text="  IFR ",command=cal_func , bg = "grey" , fg = "blue")
button1.grid(row=8,column=1)

window.mainloop()
```
